```python
def model_selection(algo_instance,x_train,y_train,x_test,y_test):
    algo_instance.fit(x_train,y_train)
    model_pred_train = algo_instance.predict(x_train)
    model_pred_test = algo_instance.predict(x_test)
    print("accuracy of training model:",accuracy_score(y_train,model_pred_train))
    print("accuracy of test data:",accuracy_score(y_test,model_pred_test))
    print("classification report of training data:",classification_report(y_train,model_pred_train))
    print("classification report of test data:",classification_report(y_test,model_pred_test))
    print("confusion matrix:",confusion_matrix(y_test,model_pred_test))
  
  
  
  
```
